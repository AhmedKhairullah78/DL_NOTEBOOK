# open vidow

```python
import cv2
import datetime
cap=cv2.VideoCapture(0)
fourcc=cv2.VideoWriter_fourcc(*'XVID')
outi=cv2.VideoWriter('outi.avi',fourcc,20.0,(640,480))
```

```python
print(cap.isOpened())
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
        
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    
    #frame=cv2.flip(frame,0)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    outi.write(frame)  
    cv2.imshow('frame',gray)
    if cv2.waitKey(1)==ord('q'):
        break
cap.release()
outi.release()
cv2.destroyAllWindows()
```

# set width , height

```python
import cv2
cap=cv2.VideoCapture(0)
print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
cap.set(3,10000)
cap.set(4,10000)

print(cap.get(3))
print(cap.get(4))
while cap.isOpened():
    ret,frame=cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Display the resulting frame
    cv2.imshow('frame', gray)
    if cv2.waitKey(1) == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# data time

```python
import cv2
import datetime
```

```python
import cv2
cap=cv2.VideoCapture(0)
fourcc=cv2.VideoWriter_fourcc(*'XVID')
out=cv2.VideoWriter('out.avi',fourcc,20.0,(640,480))
```

```python
print(cap.isOpened())
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
        
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    
    #frame=cv2.flip(frame,0)
    font=cv2.FONT_HERSHEY_SIMPLEX
    text='width: '+str(cap.get(3))+' height: '+str(cap.get(4))
    date=str(datetime.datetime.now())
    frame=cv2.putText(frame,text,(10,100),font,1,(0,255,255),2)
    frame=cv2.putText(frame,date,(10,450),font,1,(0,255,255),2)
    #gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    out.write(frame)  
    cv2.imshow('frame',frame)
    if cv2.waitKey(1)==ord('q'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# mouse events

```python
import numpy as np
import cv2
```

```python
ev=[i for i in dir(cv2) if 'EVENT' in i]
print (ev)
```

```python
def click(ev,x,y,flags,param):
        if ev==cv2.EVENT_LBUTTONDOWN:
                print(x,' , ',y)
                font=cv2.FONT_HERSHEY_SIMPLEX
                strxy=str(x)+' , '+str(y)
                cv2.putText(img,strxy,(x,y),font, .5,(255,255,0),2)
                cv2.imshow('image',img)
        if ev==cv2.EVENT_RBUTTONDOWN:
                blue=img[y,x,0]
                green=img[y,x,1]
                red=img[y,x,2]
                font=cv2.FONT_HERSHEY_SIMPLEX 
                strBGR='[ '+str(blue)+' , '+str(green)+' , '+str(red)+' ]'
                cv2.putText(img,strBGR,(x,y),font, .5,(0,245,255),2)
                cv2.imshow('image',img)

```

```python
img=cv2.imread('unnamed.jpg')
cv2.imshow('image',img)

cv2.setMouseCallback('image',click)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
